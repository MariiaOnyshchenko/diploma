```python
import spacy
import pymorphy3
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import nltk
import streamlit as st
from openpyxl import Workbook
from openpyxl.styles import PatternFill
import uk_core_news_trf
```

```python
# Завантаження моделей NLP
nlp = spacy.load("uk_core_news_trf")
morph = pymorphy3.MorphAnalyzer(lang='uk')
with open('stopwords_ua.txt') as f:
    stopwords = set(f.read().splitlines())

# Завантаження словника тональності з Excel
ex_data = pd.read_excel('DictCopy.xlsx')
sentiment_dict = dict(zip(ex_data['Слово//словосполучення'], ex_data['Середня оцінка тональності']))
intensifiers = {
    "дуже": 1.5, "сильно": 2, "вкрай": 2.5, "надзвичайно": 3, "особливо": 2, "цілком": 1.8, "величезно": 2.2, "надмірно": 2.5,
    "неймовірно": 3, "екстремально": 3, "вкрай сильно": 3.5, "значно": 1.7
}
negations = {
    "не", "ні", "ані"
}

# Удосконалена функція попередньої обробки тексту
def preprocess_text(text):
    doc = nlp(text)
    processed_tokens = [
        morph.normal_forms(token.text)[0]
        for token in doc if token.text.lower() not in stopwords and token.is_alpha
    ]
    return processed_tokens


# Удосконалений розрахунок тональності з використанням складнішого синтаксичного аналізу
def compute_sentiment_syntax(sentence):
    doc = nlp(sentence)
    sentiment_score = 0
    word_sentiments = {}
    word_details = {}

    # Перший прохід: визначення всіх заперечень та інтенсифікаторів
    negation_scope = {}
    intensifier_scope = {}

    # Фраза для відладки
    # debug_info = []

    # Перший прохід для виявлення всіх негаторів і інтенсифікаторів
    for token in doc:
        # Пропускаємо пунктуацію
        if token.is_punct:
            continue

        # Відстежуємо область заперечення
        if token.lemma_.lower() in negations:
            # Визначаємо область дії заперечення
            # Знаходимо головне слово або наступне змістовне слово
            target_tokens = []

            # 1. Спочатку перевіряємо, чи є заперечення модифікатором для дієслова або іншого слова
            if token.head.i != token.i:  # Якщо не є коренем
                target_tokens.append(token.head.i)

                # Додаємо також всі дочірні елементи головного слова
                for child in token.head.children:
                    if child.i != token.i and not child.is_punct:
                        target_tokens.append(child.i)

            # 2. Додаємо наступні змістовні слова в області дії
            next_words = [t.i for t in doc[token.i+1:] if not t.is_punct
                         and t.pos_ not in ['CCONJ', 'SCONJ', 'ADP', 'DET']
                         and t.i - token.i <= 5]  # Максимальна відстань в 5 токенів

            if next_words:
                target_tokens.extend(next_words[:2])  # Беремо лише перші два слова

            # Застосовуємо заперечення до всіх цільових токенів
            for target_idx in target_tokens:
                negation_scope[target_idx] = True
                # debug_info.append(f"Заперечення '{token.text}' впливає на '{doc[target_idx].text}'")

        # Відстежуємо область інтенсифікатора
        if token.lemma_.lower() in intensifiers:
            intensity_value = intensifiers[token.lemma_.lower()]
            target_tokens = []

            # 1. Якщо інтенсифікатор модифікує заперечення, ми хочемо підсилити
            # дію заперечення на наступне слово
            next_token_idx = token.i + 1
            while next_token_idx < len(doc) and doc[next_token_idx].is_punct:
                next_token_idx += 1

            if next_token_idx < len(doc) and doc[next_token_idx].lemma_.lower() in negations:
                # Шукаємо слова, на які впливає заперечення
                for idx, is_negated in negation_scope.items():
                    if is_negated and idx > next_token_idx:
                        target_tokens.append(idx)
                        # debug_info.append(f"Інтенсифікатор '{token.text}' підсилює заперечення для '{doc[idx].text}'")

            # 2. Якщо після інтенсифікатора немає заперечення, або додатково потрібно
            # обробити інші слова
            if not target_tokens:
                # Перевіряємо, чи є інтенсифікатор модифікатором
                if token.head.i != token.i:  # Якщо не є коренем
                    target_tokens.append(token.head.i)
                    # debug_info.append(f"Інтенсифікатор '{token.text}' підсилює '{doc[token.head.i].text}'")

                # Якщо немає головного слова, шукаємо наступне змістовне слово
                if not target_tokens:
                    for next_word in doc[token.i+1:]:
                        if not next_word.is_punct and next_word.pos_ not in ['CCONJ', 'SCONJ', 'ADP', 'DET']:
                            target_tokens.append(next_word.i)
                            # debug_info.append(f"Інтенсифікатор '{token.text}' підсилює наступне слово '{next_word.text}'")
                            break

            # Застосовуємо інтенсифікацію до всіх цільових токенів
            for target_idx in target_tokens:
                intensifier_scope[target_idx] = intensifier_scope.get(target_idx, 1) * intensity_value

    # Другий прохід: обчислення тональності з урахуванням контексту
    for token in doc:
        # Пропускаємо пунктуацію
        if token.is_punct:
            continue

        word = token.lemma_.lower()
        sentiment_value = sentiment_dict.get(word, 0)

        # Якщо слово не має тональності і немає модифікаторів, пропускаємо його
        if sentiment_value == 0 and token.i not in negation_scope and token.i not in intensifier_scope:
            continue

        # Застосовуємо заперечення якщо воно в області дії
        negation_factor = -1 if negation_scope.get(token.i, False) else 1

        # Застосовуємо інтенсифікацію якщо вона в області дії
        intensity_factor = intensifier_scope.get(token.i, 1)

        # Обчислюємо кінцеву тональність цього слова
        final_sentiment = sentiment_value * negation_factor * intensity_factor

        # Зберігаємо деталі для візуалізації
        word_sentiments[token.i] = final_sentiment
        word_details[token.i] = {
            'text': token.text,
            'lemma': word,
            'pos': token.pos_,
            'dep': token.dep_,
            'base_sentiment': sentiment_value,
            'negated': negation_scope.get(token.i, False),
            'intensified': intensity_factor > 1,
            'intensity_factor': intensity_factor,
            'final_sentiment': final_sentiment,
        }

        # Додаємо до загальної оцінки (слова, ближчі до кореня, мають більшу вагу)
        distance_to_root = len(list(token.ancestors))
        weight = 1.0 / (1.0 + 0.1 * distance_to_root)  # Слова, ближчі до кореня, мають більше значення
        sentiment_score += final_sentiment * weight

    # Для відладки можна вивести інформацію про вплив модифікаторів
    # print("\n".join(debug_info))

    return sentiment_score, word_sentiments, word_details




# Покращена візуалізація синтаксичного дерева з додатковою інформацією
def visualize_syntax_tree(sentence, save_path="syntax_tree.png"):
    doc = nlp(sentence)
    sentiment_score, word_sentiments, word_details = compute_sentiment_syntax(sentence)

    # Виводимо детальну інформацію про аналіз тональності
    print(f"Речення: {sentence}")
    print(f"Загальна оцінка тональності: {sentiment_score:.2f}")
    print("\nДеталі аналізу окремих слів:")

    for i, token in enumerate(doc):
        if i in word_details:
            details = word_details[i]
            status = []
            if details['negated']:
                status.append("заперечення")
            if details['intensified']:
                status.append(f"інтенсифікація (×{details['intensity_factor']:.1f})")

            status_str = f" [{', '.join(status)}]" if status else ""
            print(f"- {token.text}: базова={details['base_sentiment']:.2f}, фінальна={details['final_sentiment']:.2f}{status_str}")

    # Створюємо спрямований граф тільки з токенів, які не є пунктуацією
    G = nx.DiGraph()

    # Мапа для відстеження оригінальних індексів до нових індексів вузлів (пропускаючи пунктуацію)
    idx_map = {}
    new_idx = 0

    # Додаємо вузли для кожного токена з інформацією про позицію (виключаючи пунктуацію)
    for i, token in enumerate(doc):
        if not token.is_punct:
            sentiment = word_sentiments.get(i, 0)
            details = word_details.get(i, {})

            # Створюємо більш інформативну мітку
            if abs(sentiment) > 0:
                label = f"{token.text}\n({sentiment:.2f})"
                if details.get('negated', False):
                    label += "\n[заперечення]"
                if details.get('intensified', False):
                    label += f"\n[×{details.get('intensity_factor', 1):.1f}]"
            else:
                label = token.text

            # Використовуємо індекс токена як унікальний ідентифікатор
            G.add_node(new_idx,
                       label=label,
                       word=token.text,
                       original_idx=i,
                       pos=token.pos_,
                       dep=token.dep_,
                       sentiment=sentiment)

            idx_map[i] = new_idx
            new_idx += 1

    # Додаємо ребра на основі залежностей відношень (виключаючи пунктуацію)
    for i, token in enumerate(doc):
        if token.is_punct:
            continue

        # Отримуємо індекс головного токена
        head_idx = token.head.i

        # Якщо головний токен є пунктуацією, знаходимо його головний токен
        current_head = token.head
        while current_head.is_punct and current_head.i != current_head.head.i:
            current_head = current_head.head
        head_idx = current_head.i

        # Додаємо ребро від головного токена до залежного (якщо не петля і головний токен не є пунктуацією)
        if head_idx != i and head_idx in idx_map:
            G.add_edge(idx_map[head_idx], idx_map[i], relation=token.dep_)

    # Знаходимо кореневий вузол
    root_token_idx = next((i for i, token in enumerate(doc) if token.dep_ == "ROOT" and not token.is_punct), 0)
    if root_token_idx in idx_map:
        root = idx_map[root_token_idx]
    else:
        # Якщо не знайдено кореневий вузол без пунктуації, використовуємо перший вузол
        root = 0 if len(G.nodes) > 0 else None

    if root is None or len(G.nodes) == 0:
        plt.figure(figsize=(8, 6))
        plt.text(0.5, 0.5, "Немає дійсного синтаксичного дерева (тільки пунктуація або порожній ввід)",
                 horizontalalignment='center', verticalalignment='center')
        plt.axis('off')
        plt.savefig(save_path)
        plt.close()
        return

    # Використовуємо горизонтальне дерево
    try:
        # Спробуємо використати graphviz для кращого макету
        import pygraphviz
        pos = nx.nx_agraph.graphviz_layout(G, prog="dot", root=root)
    except ImportError:
        try:
            # Альтернатива: використовуємо pydot
            import pydot
            pos = nx.drawing.nx_pydot.graphviz_layout(G, prog="dot", root=root)
        except ImportError:
            # Повертаємося до вбудованого макету networkx
            pos = hierarchy_pos(G, root, width=2*len(G.nodes), xcenter=0)

    # Коригуємо позиції, щоб зробити дерево горизонтальним
    pos = {k: (y, x) for k, (x, y) in pos.items()}

    # Витягуємо мітки вузлів
    labels = {node: data["label"] for node, data in G.nodes(data=True)}

    # Малюємо граф
    plt.figure(figsize=(16, 10))

    # Колір вузлів залежно від тональності та частини мови
    node_colors = []
    node_sizes = []
    node_shapes = []

    for i in G.nodes():
        sentiment = G.nodes[i]['sentiment']
        pos_tag = G.nodes[i]['pos']

        # Розмір залежно від важливості (ROOT та основні змістовні слова більші)
        if G.nodes[i]['dep'] == 'ROOT':
            node_sizes.append(4000)
        elif pos_tag in ['NOUN', 'VERB', 'ADJ', 'ADV']:
            node_sizes.append(3000)
        else:
            node_sizes.append(2000)

        # Колір залежно від тональності
        if sentiment > 0:
            intensity = min(1.0, abs(sentiment) / 3.0)
            node_colors.append((0.7 - 0.5 * intensity, 0.9, 0.7 - 0.5 * intensity))  # Зелений з інтенсивністю
        elif sentiment < 0:
            intensity = min(1.0, abs(sentiment) / 3.0)
            node_colors.append((0.9, 0.7 - 0.5 * intensity, 0.7 - 0.5 * intensity))  # Червоний з інтенсивністю
        else:
            if pos_tag in ['NOUN', 'PROPN']:
                node_colors.append('lightblue')
            elif pos_tag == 'VERB':
                node_colors.append('lightyellow')
            elif pos_tag in ['ADJ', 'ADV']:
                node_colors.append('lightcyan')
            else:
                node_colors.append('lightgray')

    # Малюємо вузли
    nx.draw_networkx_nodes(G, pos,
                          node_color=node_colors,
                          node_size=node_sizes,
                          alpha=0.9)

    # Малюємо ребра зі стрілками та власним стилем
    for edge in G.edges():
        # Отримуємо тип відношення
        relation = G[edge[0]][edge[1]]['relation']

        # Власний стиль стрілки залежно від відношення
        if relation in ['obj', 'dobj', 'iobj']:
            style = 'solid'
            width = 2.0
            arrow_style = '->'
        elif relation in ['nsubj', 'csubj']:
            style = 'solid'
            width = 2.0
            arrow_style = '->'
        elif relation in ['amod', 'advmod']:
            style = 'dotted'
            width = 1.5
            arrow_style = '->'
        else:
            style = 'dashed'
            width = 1.0
            arrow_style = '->'

        nx.draw_networkx_edges(G, pos,
                              edgelist=[edge],
                              width=width,
                              style=style,
                              alpha=0.7,
                              arrows=True,
                              arrowstyle=arrow_style,
                              arrowsize=15)

    # Малюємо мітки вузлів
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=10, font_weight='bold')

    # Додаємо мітки ребер для залежностей відношень
    edge_labels = {(u, v): G[u][v]["relation"] for u, v in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

    # Додаємо заголовок та легенду
    plt.title(f"Дерево синтаксичних залежностей з тональністю: '{sentence}'\nЗагальна оцінка: {sentiment_score:.2f}", fontsize=16)

    # Додаємо легенду для типів вузлів
    legend_elements = [
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='lightblue', markersize=10, label='Іменник'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='lightyellow', markersize=10, label='Дієслово'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='lightcyan', markersize=10, label='Прикметник/Прислівник'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='lightgray', markersize=10, label='Інше'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=(0.4, 0.9, 0.4), markersize=10, label='Позитивна тональність'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=(0.9, 0.4, 0.4), markersize=10, label='Негативна тональність')
    ]
    plt.legend(handles=legend_elements, loc='upper right')

    # Видаляємо осі
    plt.axis('off')

    # Зберігаємо та показуємо
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()

# Допоміжна функція для ієрархічного макету без зовнішніх залежностей
def hierarchy_pos(G, root=None, width=1., vert_gap=0.2, vert_loc=0, xcenter=0.5):
    """
    Власний алгоритм ієрархічного позиціонування для макету дерева
    """
    if root is None:
        if isinstance(G, nx.DiGraph):
            root = next(iter(nx.topological_sort(G)))  # Корінь - перший вузол у топологічному сортуванні
        else:
            root = np.random.choice(list(G.nodes))

    def _hierarchy_pos(G, root, width=1., vert_gap=0.2, vert_loc=0, xcenter=0.5, pos=None, parent=None, parsed=None):
        if pos is None:
            pos = {root: (xcenter, vert_loc)}
        else:
            pos[root] = (xcenter, vert_loc)

        children = list(G.neighbors(root))
        if parent is not None:
            children.remove(parent)  # Видаляємо батьківський вузол, щоб запобігти циклічному BFS
        if not children:
            return pos

        dx = width / len(children)
        nextx = xcenter - width/2 + dx/2
        for child in children:
            pos = _hierarchy_pos(G, child, width=dx, vert_gap=vert_gap,
                                vert_loc=vert_loc-vert_gap, xcenter=nextx,
                                pos=pos, parent=root, parsed=parsed)
            nextx += dx
        return pos

    return _hierarchy_pos(G, root, width=width, vert_gap=vert_gap, vert_loc=vert_loc, xcenter=xcenter)

# Експорт до Excel
def export_to_excel(sentiment_scores):
    wb = Workbook()
    ws = wb.active
    for row, sentiment in enumerate(sentiment_scores, start=1):
        cell = ws.cell(row=row, column=2, value=sentiment)
        if sentiment > 0:
            cell.fill = PatternFill(start_color="00FF00", fill_type="solid")
        elif sentiment < 0:
            cell.fill = PatternFill(start_color="FF0000", fill_type="solid")
    wb.save("sentiment_results.xlsx")

if __name__ == '__main__':
    # Демонстрація роботи з різними прикладами
    test_sentences = [
        "Я вкрай не люблю запізнюватись!",
        "Сьогодні дуже гарна погода.",
        "Мені не подобається цей фільм.",
        "Я надзвичайно сильно не рекомендую цей ресторан.",
        "Це не дуже цікава книга.",
        "Я люблю і радію, але і сумую."
    ]

    for sentence in test_sentences:
        print("\n" + "="*60)
        sentiment_score, _, _ = compute_sentiment_syntax(sentence)
        print(f"Тестове речення: {sentence}")
        print(f"Оцінка тональності: {sentiment_score:.2f}")

    # Інтерактивний режим
    print("\n" + "="*60)
    text_input = input("\nВведіть речення для аналізу: ")

    if text_input:
        sentiment_score, _, _ = compute_sentiment_syntax(text_input)
        print(f"Оцінка тональності: {sentiment_score:.2f}")
        visualize_syntax_tree(text_input)
```

```python
# Імпорт необхідних бібліотек
import pandas as pd
import spacy
import pymorphy3
# import networkx as nx # Removed
# import matplotlib.pyplot as plt # Removed
# import numpy as np # Removed
from openpyxl import Workbook
from openpyxl.styles import PatternFill
import os
import zipfile
from pathlib import Path
import warnings
# import matplotlib.font_manager as fm # Removed
warnings.filterwarnings("ignore")

# Налаштування matplotlib для кращого відображення українського тексту - Removed
# available_fonts = [f.name for f in fm.fontManager.ttflist] # Removed
# if 'Arial Unicode MS' in available_fonts: # Removed
#     plt.rcParams['font.family'] = ['Arial Unicode MS'] # Removed
# elif 'DejaVu Sans' in available_fonts: # Removed
#     plt.rcParams['font.family'] = ['DejaVu Sans'] # Removed
# else: # Removed
#     plt.rcParams['font.family'] = 'sans-serif' # Removed
# plt.rcParams['axes.unicode_minus'] = False # Removed

print("Loading NLP models...")

# Завантаження моделей NLP
nlp = spacy.load("uk_core_news_trf")
morph = pymorphy3.MorphAnalyzer(lang='uk')
with open('stopwords_ua.txt') as f:
    stopwords = set(f.read().splitlines())

# Завантаження словника тональності з Excel
ex_data = pd.read_excel('DictCopy.xlsx')
sentiment_dict = dict(zip(ex_data['Слово//словосполучення'], ex_data['Середня оцінка тональності']))
intensifiers = {
    "дуже": 1.5, "сильно": 2, "вкрай": 2.5, "надзвичайно": 3, "особливо": 2, "цілком": 1.8, "величезно": 2.2, "надмірно": 2.5,
    "неймовірно": 3, "екстремально": 3, "вкрай сильно": 3.5, "значно": 1.7
}
negations = {
    "не", "ні", "ані"
}

# Удосконалена функція попередньої обробки тексту
def preprocess_text(text):
    doc = nlp(text)
    processed_tokens = [
        morph.normal_forms(token.text)[0]
        for token in doc if token.text.lower() not in stopwords and token.is_alpha
    ]
    return processed_tokens

# Удосконалений розрахунок тональності з використанням складнішого синтаксичного аналізу
def compute_sentiment_syntax(sentence):
    doc = nlp(sentence)
    sentiment_score = 0
    word_sentiments = {} # Still computed, but not used by removed tree functions
    word_details = {}    # Still computed, but not used by removed tree functions

    # Перший прохід: визначення всіх заперечень та інтенсифікаторів
    negation_scope = {}
    intensifier_scope = {}

    # Перший прохід для виявлення всіх негаторів і інтенсифікаторів
    for token in doc:
        # Пропускаємо пунктуацію
        if token.is_punct:
            continue

        # Відстежуємо область заперечення
        if token.lemma_.lower() in negations:
            target_tokens = []
            if token.head.i != token.i:  # Якщо не є коренем
                target_tokens.append(token.head.i)
                for child in token.head.children:
                    if child.i != token.i and not child.is_punct:
                        target_tokens.append(child.i)
            next_words = [t.i for t in doc[token.i+1:] if not t.is_punct
                          and t.pos_ not in ['CCONJ', 'SCONJ', 'ADP', 'DET']
                          and t.i - token.i <= 5]  # Максимальна відстань в 5 токенів
            if next_words:
                target_tokens.extend(next_words[:2])  # Беремо лише перші два слова
            for target_idx in target_tokens:
                negation_scope[target_idx] = True

        # Відстежуємо область інтенсифікатора
        if token.lemma_.lower() in intensifiers:
            intensity_value = intensifiers[token.lemma_.lower()]
            target_tokens = []
            next_token_idx = token.i + 1
            while next_token_idx < len(doc) and doc[next_token_idx].is_punct:
                next_token_idx += 1
            if next_token_idx < len(doc) and doc[next_token_idx].lemma_.lower() in negations:
                for idx, is_negated in negation_scope.items():
                    if is_negated and idx > next_token_idx:
                        target_tokens.append(idx)
            if not target_tokens:
                if token.head.i != token.i:  # Якщо не є коренем
                    target_tokens.append(token.head.i)
                if not target_tokens:
                    for next_word in doc[token.i+1:]:
                        if not next_word.is_punct and next_word.pos_ not in ['CCONJ', 'SCONJ', 'ADP', 'DET']:
                            target_tokens.append(next_word.i)
                            break
            for target_idx in target_tokens:
                intensifier_scope[target_idx] = intensifier_scope.get(target_idx, 1) * intensity_value

    # Другий прохід: обчислення тональності з урахуванням контексту
    for token in doc:
        if token.is_punct:
            continue
        word = token.lemma_.lower()
        sentiment_value = sentiment_dict.get(word, 0)
        if sentiment_value == 0 and token.i not in negation_scope and token.i not in intensifier_scope:
            continue
        negation_factor = -1 if negation_scope.get(token.i, False) else 1
        intensity_factor = intensifier_scope.get(token.i, 1)
        final_sentiment = sentiment_value * negation_factor * intensity_factor
        word_sentiments[token.i] = final_sentiment
        word_details[token.i] = {
            'text': token.text,
            'lemma': word,
            'pos': token.pos_,
            'dep': token.dep_,
            'base_sentiment': sentiment_value,
            'negated': negation_scope.get(token.i, False),
            'intensified': intensity_factor > 1,
            'intensity_factor': intensity_factor,
            'final_sentiment': final_sentiment,
        }
        distance_to_root = len(list(token.ancestors))
        weight = 1.0 / (1.0 + 0.1 * distance_to_root)
        sentiment_score += final_sentiment * weight

    return sentiment_score, word_sentiments, word_details

def compute_sentiment_for_text(text):
    """
    Обчислює тональність для всього тексту, що може мати кілька речень.
    Якщо більше ніж 1 речення — ділимо результат на кількість речень.
    """
    doc = nlp(text)
    sentences = list(doc.sents)

    total_sentiment = 0.0
    sentence_count = 0
    # combined_word_sentiments and combined_word_details are computed but not used by subsequent non-tree logic
    combined_word_sentiments = {}
    combined_word_details = {}

    for sent in sentences:
        sentence_text = sent.text.strip()
        if not sentence_text:
            continue
        sentiment, word_sentiments, word_details = compute_sentiment_syntax(sentence_text)
        total_sentiment += sentiment
        sentence_count += 1
        combined_word_sentiments.update(word_sentiments)
        combined_word_details.update(word_details)

    if sentence_count > 1:
        total_sentiment /= sentence_count

    return total_sentiment, combined_word_sentiments, combined_word_details

# Removed visualize_syntax_tree function
# Removed hierarchy_pos function

def compare_ratings(sentiment_score, stars):
    """
    Порівнює оцінку програми з зірками користувача
    """
    if sentiment_score == 0:
        return "no match"

    # Негативна тональність і 1 зірка
    if sentiment_score <= 2 and stars == 1:
        return "match"

    # Позитивна тональність і 5 зірок
    if sentiment_score > 2 and stars in [3, 4, 5]:
        return "match"

    return "no match"

# Removed create_sentiment_trees_folder function
# Removed save_syntax_tree_png function


def process_csv_reviews(csv_file_path, batch_size=100): # Removed save_trees parameter
    """
    Обробляє CSV файл з оглядами
    batch_size: розмір пакету для збереження проміжних результатів
    """
    print("Завантажуємо CSV файл...")
    df = pd.read_csv(csv_file_path)

    required_columns = ['cleaned_text', 'stars']
    for col in required_columns:
        if col not in df.columns:
            raise ValueError(f"Колонка '{col}' не знайдена в CSV файлі")

    print(f"Знайдено {len(df)} оглядів для обробки")

    # Removed tree folder creation

    sentiment_scores = []
    comparison_results = []

    processed_count = 0
    error_count = 0

    for idx, row in df.iterrows():
        cleaned_text = str(row['cleaned_text'])
        stars = row['stars']

        print(f"Обробка огляду {idx + 1}/{len(df)} - {cleaned_text[:50]}...")

        try:
            sentiment_score, _, _ = compute_sentiment_for_text(cleaned_text) # word_sentiments, word_details ignored
            sentiment_scores.append(sentiment_score)

            comparison = compare_ratings(sentiment_score, stars)
            comparison_results.append(comparison)

            # Removed tree saving logic

            processed_count += 1

            if processed_count % batch_size == 0:
                print(f"Збереження проміжних результатів... ({processed_count}/{len(df)})")
                temp_df = df.iloc[:processed_count].copy() # Ensure we are working with a copy for slices
                # Ensure sentiment_scores and comparison_results align with the slice
                temp_df['rate by program'] = sentiment_scores[:processed_count]
                temp_df['match/nomatch'] = comparison_results[:processed_count]
                temp_df.to_csv(f'temp_results_{processed_count}.csv', index=False)
                print(f"✅ Збережено проміжні результати для {processed_count} оглядів")

        except Exception as e:
            print(f"❌ Критична помилка для огляду {idx + 1}: {e}")
            sentiment_scores.append(0.0)
            comparison_results.append("error")
            error_count += 1

    # Ensure the lists are of the same length as the DataFrame before assignment
    # This handles cases where the loop might exit prematurely or errors occur
    # However, the current loop structure appends for every row, so lengths should match
    # if df was not modified mid-loop. For safety, one might pad if necessary,
    # but the current logic seems to maintain length.

    df['rate by program'] = sentiment_scores
    df['match/nomatch'] = comparison_results

    output_file = 'vidhuk_reviews_with_sentiment.csv'
    df.to_csv(output_file, index=False)
    print(f"Результати збережено в файл: {output_file}")
    print(f"Обробка завершена: {processed_count} успішно, {error_count} помилок")

    import glob
    for temp_file in glob.glob('temp_results_*.csv'):
        try:
            os.remove(temp_file)
            print(f"Видалено тимчасовий файл: {temp_file}")
        except:
            pass

    create_analysis_report(df)

    return df

def create_analysis_report(df):
    """
    Створює звіт з аналізу тональності
    """
    print("\n" + "="*60)
    print("ЗВІТ З АНАЛІЗУ ТОНАЛЬНОСТІ")
    print("="*60)

    total_reviews = len(df)
    matches = len(df[df['match/nomatch'] == 'match'])
    no_matches = len(df[df['match/nomatch'] == 'no match'])

    print(f"Загальна кількість оглядів: {total_reviews}")
    if total_reviews > 0:
        print(f"Збігів (match): {matches} ({matches/total_reviews*100:.1f}%)")
        print(f"Незбігів (no match): {no_matches} ({no_matches/total_reviews*100:.1f}%)")
    else:
        print("Збігів (match): 0 (0.0%)")
        print("Незбігів (no match): 0 (0.0%)")

    positive_reviews = len(df[df['rate by program'] > 0])
    negative_reviews = len(df[df['rate by program'] < 0])
    neutral_reviews = len(df[df['rate by program'] == 0])

    print(f"\nРозподіл за тональністю:")
    if total_reviews > 0:
        print(f"Позитивні огляди: {positive_reviews} ({positive_reviews/total_reviews*100:.1f}%)")
        print(f"Негативні огляди: {negative_reviews} ({negative_reviews/total_reviews*100:.1f}%)")
        print(f"Нейтральні огляди: {neutral_reviews} ({neutral_reviews/total_reviews*100:.1f}%)")
    else:
        print("Позитивні огляди: 0 (0.0%)")
        print("Негативні огляди: 0 (0.0%)")
        print("Нейтральні огляди: 0 (0.0%)")

    print(f"\nРозподіл за зірками:")
    if 'stars' in df.columns and not df['stars'].empty:
        for stars_val in sorted(df['stars'].unique()):
            count = len(df[df['stars'] == stars_val])
            if total_reviews > 0:
                 print(f"{stars_val} зірок: {count} ({count/total_reviews*100:.1f}%)")
            else:
                print(f"{stars_val} зірок: {count} (0.0%)")
    else:
        print("Немає даних по зіркам для відображення розподілу.")

    print("\n" + "="*60)

if __name__ == '__main__':
    csv_file_path = 'vidhuk_reviews_cleaned_extra_5.csv'

    # Removed SAVE_TREES variable
    BATCH_SIZE = 100

    try:
        print("Починаємо обробку CSV файлу...")
        # Removed save_trees from print
        print(f"Налаштування: розмір пакету = {BATCH_SIZE}")

        df_results = process_csv_reviews(csv_file_path,
                                           batch_size=BATCH_SIZE) # Removed save_trees argument

        print("\nОбробка завершена успішно!")
        # Removed print statement about saving trees
        print(f"Результати збережено в файлі 'vidhuk_reviews_with_sentiment.csv'")

    except FileNotFoundError:
        print(f"Файл '{csv_file_path}' не знайдено. Переконайтеся, що файл існує.")
    except Exception as e:
        print(f"Помилка під час обробки: {e}")
        # Removed save_trees from print
        print("Спробуйте зменшити BATCH_SIZE")

    print("\n" + "="*60)
    print("ДЕМОНСТРАЦІЯ РОБОТИ АЛГОРИТМУ")
    print("="*60)

    test_sentences = [
        "Я вкрай не люблю запізнюватись!",
        "Сьогодні дуже гарна погода.",
        "Мені не подобається цей фільм.",
        "Я надзвичайно сильно не рекомендую цей ресторан.",
        "Це не дуже цікава книга.",
        "Я люблю і радію, але і сумую."
    ]

    for sentence in test_sentences:
        print("\n" + "-"*40)
        # The compute_sentiment_syntax function still returns word_sentiments and word_details,
        # but they are ignored here as they were primarily for the tree visualization.
        sentiment_score, _, _ = compute_sentiment_syntax(sentence)
        print(f"Тестове речення: {sentence}")
        print(f"Оцінка тональності: {sentiment_score:.2f}")
```
